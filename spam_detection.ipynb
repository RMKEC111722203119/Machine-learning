```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer

```

```python
data = pd.read_csv('/content/spam.csv',encoding='ISO-8859-1')
```

```python
data.head()
```

```python
co= [name for name in data.columns if name.startswith('Unnamed')]
data.drop(columns=co,inplace=True)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
info = data.where((pd.notnull(data)),'')
```

```python
info.shape
```

```python
info.rename(columns=dict({"v1":"target","v2":"message"}),inplace=True)
```

```python
info.loc[info['target']=='spam','target',]=0
info.loc[info['target']=='ham','target',]=1
```

```python
info.head()
```

```python
x = info["message"]

y= info["target"]
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
fe = TfidfVectorizer(min_df=1,stop_words="english",lowercase=True)
```

```python
trx=fe.fit_transform(x_train)
tex=fe.transform(x_test)
```

```python
y_train=y_train.astype(int)
y_test=y_test.astype(int)
```

```python
model = LogisticRegression()
```

```python
model.fit(trx,y_train)
```

```python
pr_td=model.predict(trx)
acc_td= accuracy_score(y_train,pr_td)

pr_td2=model.predict(tex)
acc_td2= accuracy_score(y_test,pr_td2)

print("accuracy on training data : ",acc_td)
print("accuracy on test data : ",acc_td2)
```

```python
def pre():
  a=input()
  input_mail = [a]
  idf= fe.transform(input_mail)
  prediction = model.predict(idf)
  if(prediction == 0):
    print("spam")
  else:
    print ("not a spam")
```

```python
while(True):
  print("""PRESS 1 FOR PREDICTION
PRESS 2 FOR EXIT""")
  A=int(input())
  if(A==2):
      break
  else:
      pre()

```
