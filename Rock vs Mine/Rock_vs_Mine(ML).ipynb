```
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```
sd=pd.read_csv("/content/sonar data (1).csv",header=None)
```

```
sd.tail()
```

```
sd.describe()
```

```
sd[60].value_counts()
```

```
x=sd.drop(columns=60,axis=1)
y=sd[60]
```

```
print(x)
```

```
print(y)
```

```
xtr,xte,ytr,yte=train_test_split(x, y, test_size = 0.1, stratify=y, random_state=2)

```

```
m=LogisticRegression()
```

```
m.fit(xtr,ytr)
```

**Model** **accuracy**

```
xtrp=m.predict(xtr)
accu=accuracy_score(xtrp,ytr)
print(accu)
```

```
xtep=m.predict(xte)
teaccu=accuracy_score(xtep,yte)
print(teaccu)
```
