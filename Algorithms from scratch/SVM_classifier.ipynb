```python
import numpy as np
```

```python
class svm_classifier():

  def __init__(self,no_of_iterations,learning_rate,lambd):

    self.no_of_iterations=no_of_iterations
    self.learning_rate=learning_rate
    self.lambd=lambd


  def fit(self,X,Y):

    self.m,self.n=X.shape

    self.w=np.zeros(self.n)
    self.b=0
    self.X=X
    self.Y=Y

    for i in range(self.no_of_iterations):
      self.update_weights()

  def update_weights(self):

    y_label=np.where(self.Y<=0,-1,1)

    #gradient

    for index ,i in enumerate(self.X):

      condition=y_label[index]*(np.dot(i,self.w)-self.b) >=1

      if(condition==True):

           dw= 2*self.lambd*self.w
           db=0
      else:
          dw=2*self.lambd*self.w - np.dot(i,y_label[index])
          db=y_label[index]

      self.w=self.w-self.learning_rate*dw
      self.b=self.b-self.learning_rate*db


  def predict(self,X):
    output = np.dot(X,self.w)-self.b
    pl=np.sign(output)
    y_hat = np.where(pl<=1,0,1)
    return y_hat
```
