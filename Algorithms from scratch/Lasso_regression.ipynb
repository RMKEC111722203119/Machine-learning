**Lasso Regression model**

Y=wx+b

x is array therefore,

x.dot(self.w)+self.b

Gradient Descent
optimisation algo to find the minimal Loss function

w = w - a*dw

b = b - b*db

where, a is the learning rate

used for updating parameters

```python
import numpy as np
```

```python
class Lasso_Regression():

  def __init__(self,Learning_rate,no_of_iterations,lamb):

    self.Learning_rate=Learning_rate
    self.no_of_iterations=no_of_iterations
    self.lamb=lamb

  def fit(self,X,Y):

    self.m,self.n=X.shape  #n=no of features,m=training sets

    self.w= np.zeros(self.n)
    self.b= 0
    self.X=X
    self.Y=Y

    dw = np.zeros(self.n)

    for i in range (self.no_of_iterations): #implementation of gradient descendant
      self.update_weights( )



  def update_weights(self,):

    Y_prediction = self.predict(self.X)

    for i in range(self.n):
      if self.w[i]>0:

         dw[i]= (-(2*(self.X[:,i]).dot(self.Y-Y_prediction))+self.lamb)/self.m  #learning rate formula
      else:
          dw[i]=(-(2*(self.X[:,i]).dot(self.Y-Y_prediction))-self.lamb)/self.m

    db= -(2*np.sum(self.Y-Y_prediction))/self.m

    self.w= self.w-self.Learning_rate*dw #gradient descendant formula

    self.b= self.b-self.Learning_rate*db



  def predict(self,X):

    return X.dot(self.w)+self.b # y=wx+b
```
